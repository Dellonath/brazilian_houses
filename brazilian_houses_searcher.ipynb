```python
import pandas as pd
houses = pd.read_csv("dataset/houses_to_rent_v2.csv")
```

```python
houses.info()
```

```python
houses.head()
```

```python
houses.describe()
```

```python
houses.groupby('city').mean()
```

```python
houses['animal'].replace(['acept', 'not acept'], [1, 0], inplace = True)
houses['furniture'].replace(['furnished', 'not furnished'], [1, 0], inplace = True)
```

```python
houses.head()
```

```python
houses.info()
```

```python
houses.groupby('floor').count().head()
```

```python
houses['floor'].replace('-', 0, inplace = True)
houses['floor'] = pd.to_numeric(houses['floor'])
```

```python
houses.info()
```

```python
houses.describe()
```

```python
houses.dropna(inplace = True)
```

```python
houses.head()
```

```python
houses.groupby('city').describe()[['area', 'total (R$)']]
```

```python
houses['city'].value_counts()
```

```python
import matplotlib.pyplot as plt

fig1, number_houses = plt.subplots()
number_houses.pie(houses['city'].value_counts(), 
        labels=houses['city'].value_counts(), 
        autopct='%1.1f%%', 
        shadow=True, 
        startangle=45)

number_houses.set_title("Number of houses by City")

number_houses.legend(['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Porto Alegre', 'Campinas'],
          title="Regions",
          loc="center left",
          bbox_to_anchor=(1.1, 0, 0.5, 1))
plt.show()
```
